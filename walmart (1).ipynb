```python
ui = pd.read_csv('/content/Bradshaw Conversation Tracker - Sheet3.csv')
blinks = []

for iui in range(len(ui['browse'])):
  blinks.append(ui['browse'][iui])

for i in range(len(ui['links'])):
  time.sleep(random.uniform(10,20))
  html = requests.get(ui['links'][i], headers = headers, cookies = cooked).text
  soupcn = str(bs4.BeautifulSoup(html, "html.parser"))

  spl = soupcn.split('https://www.walmart.com/browse/')

  for xx in range(len(spl)-1):
    try:
      if len(spl[xx+1].split('"')[0].split('/')[1])>=3:
        blinks.append('https://www.walmart.com/browse/'+spl[xx+1].split('"')[0])
    except:
      print('error with: '+ui['links'][i])
blinks
```

```python
links = []
df = pd.DataFrame()
df['links']=pd.DataFrame(blinks).drop_duplicates()[0].reset_index().drop(columns='index')
cat = []
sub_cat = []
lnks = []
for ii in range(len(df['links'])):
  try:
    sub_cat.append(df['links'][ii].split('https://www.walmart.com/browse/')[1].split('/')[1].replace('-',' '))
    cat.append(df['links'][ii].split('https://www.walmart.com/browse/')[1].split('/')[0].replace('-',' '))
    lnks.append(df['links'][ii])
  except:
    print('nope')
dff = pd.DataFrame()
dff['sub cats'] = sub_cat

dff['broad cats'] = cat
dff['links'] = lnks
dff.to_csv('saved.csv')
```
